```python
# import pandas library to load the data set.
import pandas as pd
```

```python
# label the data set and import the .csv file.
# we are skipping the first row, and the two footers, use the engine 'pytohn'.
df = pd.read_csv('LoanStats_2019Q1.csv', skiprows=1, skipfooter=2, engine='python')
# show the shape of the data frame.
print(df.shape)
# display the headers of the data frame.
df.head()
```

## Working with strings.

```python
# look at which columns are objects, 'strings'.
df.describe(include='object')
```

```python
# transpose .T will list all the features and be easier to view.
df.describe(include='object').T
```

```python
# remove the 'months' text with str.strrip & change to integer with .astype(int).
df['term'] = df['term'].str.strip('months').astype(int)
```

```python
# check that the text is removed and 'term' is now a float.
df['term'].head()
```

```python
# look at the different 'values' for loan status column.
df['loan_status'].value_counts()
```

```python
# add an end column 'loan_status_is_great' that will show if the loan status is current/full paid as 1, and the rest as 0.
df['loan_status_is_great'] = df['loan_status'].map({'Current':1, 'Fully Paid':1, 'Late (31-120 days)':0, 'In Grace Period':0, 'Late (16-30 days)':0, 'Charged Off':0})
# show the data frame with headers.
df.head()
```

```python
# check that the new column has the right # of 0's & 1's.
df['loan_status_is_great'].value_counts()
```

```python
# check the loan_status column to confirmed the #'s are correct.
df['loan_status'].value_counts()
```

```python
# with pd.to_datetime we can adjust the issue date 'text' to #s display.
df['issue_d'].describe()
```

```python
# with pd.to_datetime we can adjust the issue date 'text' to #s display.
df['issue_d'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
```

```python
# look at the new date format we set above.
df['issue_d'].describe()
```

```python
# look at the new date format we set above.
df['last_pymnt_d'].describe()
```

```python
df['last_pymnt_d'].value_counts()
```

```python
# add an end column 'last_payment_month' that will show the month of the last payment due.
df['last_pymnt_d_m'] = df['last_pymnt_d'].map({'Jan-2019':1, 'Feb-2019':2, 'Mar-2019':3, 'Apr-2019':4, 'May-2019':5, 'Jun-2019':6, 'Jul-2019':7})
df['last_pymnt_d_m'].value_counts()
```

```python
# add an end column 'last_payment_month' that will show the month of the last payment due.
df['last_pymnt_d_y'] = df['last_pymnt_d'].map({'Jan-2019':2019, 'Feb-2019':2019, 'Mar-2019':2019, 'Apr-2019':2019, 'May-2019':2019, 'Jun-2019':2019, 'Jul-2019':2019})
df['last_pymnt_d_y'].value_counts()
```

```python
# check with columns are using a date '_d'.
[col for col in df if col.endswith('_d')]
```

```python
# now i am going to change all the columns with a date '_d' to a #s format, we already changed 'issue_d' above so we will leave that out.
for col in ['last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d']:
  df[col] = pd.to_datetime(df[col])
```

```python
# look at all the '_d' columns we used pd.to_datetime on.
df.describe(include='datetime')
```

```python
# create a new data frame of all the changes & additions we made.
df1 = df[['issue_d','term','last_pymnt_d_m','last_pymnt_d_y','loan_status_is_great']]
```
